```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
L = 10e-9  # Width of the square well in meters (10 nm)
n = 3
# ------vvv----graph range---vvv--------
x = np.linspace(-10e-9, 10e-9, 1000)
# ---vvv---wave-function---vvv---
wave_function = np.zeros_like(x)
# ---vvv---inside-infinite-square-well---vvv-------
cosine_region = (x >= -5e-9) & (x <= 5e-9)
# ---vvv---actual-wave-function---vvv---
wave_function[cosine_region] = (np.sqrt(2) / L) * np.cos(n * np.pi * x[cosine_region] / L)

# ------vvv---plot---vvv-------
plt.figure(figsize=(8, 6))
plt.plot(x, wave_function, label=f'n = {n} State')
plt.title('Wave Function for an Electron in the n = 3 State')
plt.xlabel('Position (meters)')
plt.ylabel('Wave Function Amplitude')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(-10e-9, 10e-9)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
L = 10e-9  # Width of the square well in meters (10 nm)
a = 1e9    # 1 nm^(-1)


x = np.linspace(-10e-9, 10e-9, 1000)

wave_function = np.zeros_like(x)

exp_region = x > 0
A = 2 * a ** (3/2)  
wave_function[exp_region] = A * np.exp(-a * x[exp_region])

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, wave_function, label='Wave Function')
plt.title('Wave Function')
plt.xlabel('Position (meters)')
plt.ylabel('Wave Function Amplitude')
plt.grid(True)
plt.legend()
plt.xlim(-10e-9, 10e-9)

plt.show()
```
